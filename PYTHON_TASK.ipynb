```python
numbers = [11, 35, 67, 23, 9, 34, 51, 78, 99, 11, 28, 31, 44, 59, 66, 77, 82, 93, 100, 10]

print("Maximum:", max(numbers))
print("Minimum:", min(numbers))
print("Sum:", sum(numbers))
print("Average:", sum(numbers) / len(numbers))
```

```python
numbers = [11, 35, 67, 23, 9, 34, 51, 78, 99, 11, 28, 31, 44, 59, 66, 77, 82, 93, 100, 10]

print("Even Numbers:")
for num in numbers:
    if num % 2 == 0:
        print(num, end=" ")

print("\nOdd Numbers:")
for num in numbers:
    if num % 2 != 0:
        print(num, end=" ")
```

```python
student = {
    "Name": "Mukund",
    "Age": 19,
    "Course": "B.Tech",
    "City": "Kolkata"
}

print("Keys:")
for key in student.keys():
    print(key)

print("\nValues:")
for value in student.values():
    print(value)
```

```python
def square(num):
    return num * num

number = 5
print("Square =", square(number))
```

```python
text = input("Enter a string: ")

count = 0
vowels = "aeiouAEIOU"

for char in text:
    if char in vowels:
        count += 1

print("Number of vowels:", count)
```

```python
numbers = [1, 2, 3, 2, 4, 5, 1, 6, 3, 7]
unique = []

for num in numbers:
    if num not in unique:
        unique.append(num)

print(unique)
```
